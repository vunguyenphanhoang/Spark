```python
data = sc.textFile(filename).cache()
```

```python
print('Lines: %i' % (len(data.collect())))
```

```python
numLoves = data.filter(lambda s: 'love' in s).count()
```

```python
print("Line with 'love': %i" % (numLoves))
```

```python
import re
pattern = "[\s|\S]*\d{1,}[\s|\S]*"
```

```python
numbers = data.filter(lambda s: re.match(pattern,s)).count()
```

```python
if numbers > 0 :
  print("Lines with numbers: %i" % (numbers))
  numbers_lines = data.filter(lambda s: re.match(pattern,s))
  i = 0
  for line in numbers_lines.collect():
    print(i ,':',line)
    i+=1
else:
  print("No number in this text")
```
