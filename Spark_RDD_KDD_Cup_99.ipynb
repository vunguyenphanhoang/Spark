4. To RDD mi, vi iu kin cc element trong RDD u c chui 'nornal'

```
normal_data = data.filter(lambda x: 'normal' in x)
```

5. Cho bit RDD mi c bao nhiu elements, mt bao nhiu time  m s elements ny

```
from time import time
t0 = time()
normal_count = normal_data.count()
t1 = time() - t0
print('C ', normal_count, 'normal')
print('Thi gian m:', format(round(t1,3)), 'giy')
```

6. To array cha tt c cc elements ban u, mt bao lu  thc hin

```
t0 = time()
array_data = data.collect()
t1 = time() - t0
print('thi gian thc hin:', round(t1,3))
```

7. To RDD = 1/10 RDD ban u vi cc elements ngu nhin

```
data_sample = data.sample(False, 0.1, 42)
sample_size = data_sample.count()
total_size = data.count()
print('s size ca mu: ', sample_size, ' trn tng ', total_size)
```

8. Cho bit t l element cha chui 'normal', so vi tng selement trong RDD sample, mt bao lu  thc hin?

```
sample_normal_tags = data_sample.filter(lambda x:'normal' in x)
t0 = time()
sample_normal_tags_count = sample_normal_tags.count()
tt = time() - t0

sample_normal_ratio = sample_normal_tags_count/ float(sample_size)
print('s lng mu normal: ',sample_normal_tags_count )
print('t l cha normal:',round(sample_normal_ratio,3) )
print('Trong: ', round(tt,3) , 'giy')
```

9. To RDD sample vi cc elements khng cha t 'normal', t RDD sample, s element l bao nhiu? in thng tin element u tin

```
sample_without_normal = data_sample.subtract(sample_normal_tags)
```

```
print('RDD sample khng cha k t normal:', sample_without_normal.count())
```

```
sample_without_normal.take(1)
```

10. Vi mi element l 1 chui to ra array_RDD c tch ra t chui vi du phn cch ',' .in thng tin u tin ca RDD ny

```
array_RDD = data.map(lambda x: x.split(','))
```

```
array_RDD.first()
```

11. c bao nhiu protocols v nhng protocols no c s dng

```
protocol = array_RDD.map(lambda x: x[1]).distinct()
protocol.collect()
```

```
print('protocols:', protocol.count())
```

12. C bao nhiu services v nhng services no c s dng

```
services = array_RDD.map(lambda x: x[2]).distinct()
services.collect()
```

```
print('services:', services.count())
```

13. C bao nhiu sn phm c to ra bi protocol v services

```
product = protocol.cartesian(services).collect()
```

```
product
```

```
print('s sn phm: ', len(product))
```
