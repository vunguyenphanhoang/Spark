4. Tạo RDD mới, với điều kiện các element trong RDD đều có chuỗi 'nornal'

```
normal_data = data.filter(lambda x: 'normal' in x)
```

5. Cho biết RDD mới có bao nhiêu elements, mất bao nhiêu time để đếm số elements này

```
from time import time
t0 = time()
normal_count = normal_data.count()
t1 = time() - t0
print('Có ', normal_count, 'normal')
print('Thời gian đếm:', format(round(t1,3)), 'giây')
```

6. Tạo array chứa tất cả các elements ban đầu, mất bao lâu để thực hiện

```
t0 = time()
array_data = data.collect()
t1 = time() - t0
print('thời gian thực hiện:', round(t1,3))
```

7. Tạo RDD = 1/10 RDD ban đầu với các elements ngẫu nhiên

```
data_sample = data.sample(False, 0.1, 42)
sample_size = data_sample.count()
total_size = data.count()
print('số size của mẫu: ', sample_size, ' trên tổng ', total_size)
```

8. Cho biết tỉ lệ element chứa chuỗi 'normal', so với tổng sốelement trong RDD sample, mất bao lâu để thực hiện?

```
sample_normal_tags = data_sample.filter(lambda x:'normal' in x)
t0 = time()
sample_normal_tags_count = sample_normal_tags.count()
tt = time() - t0

sample_normal_ratio = sample_normal_tags_count/ float(sample_size)
print('số lượng mẫu normal: ',sample_normal_tags_count )
print('tỉ lệ chứa normal:',round(sample_normal_ratio,3) )
print('Trong: ', round(tt,3) , 'giây')
```

9. Tạo RDD sample với các elements không chứa từ 'normal', từ RDD sample, số element là bao nhiêu? in thông tin element đầu tiên

```
sample_without_normal = data_sample.subtract(sample_normal_tags)
```

```
print('RDD sample không chứa kí tự normal:', sample_without_normal.count())
```

```
sample_without_normal.take(1)
```

10. Với mỗi element là 1 chuỗi tạo ra array_RDD được tách ra từ chuỗi với dấu phân cách ',' .in thông tin đầu tiên của RDD này

```
array_RDD = data.map(lambda x: x.split(','))
```

```
array_RDD.first()
```

11. có bao nhiêu protocols và những protocols nào được sử dụng

```
protocol = array_RDD.map(lambda x: x[1]).distinct()
protocol.collect()
```

```
print('protocols:', protocol.count())
```

12. Có bao nhiêu services và những services nào được sử dụng

```
services = array_RDD.map(lambda x: x[2]).distinct()
services.collect()
```

```
print('services:', services.count())
```

13. Có bao nhiêu sản phẩm được tạo ra bởi protocol và services

```
product = protocol.cartesian(services).collect()
```

```
product
```

```
print('số sản phẩm: ', len(product))
```
